# Data analysis 
* after the data is cleaned you can start your analysis
* useful commands:
    * `df.mean()`, `df.std()`, `df.min()`, `df.max()`: some basic stats
    * `df.describe()`: the above stats and more in one command
    * `df.plot()`: plot all columns (using index as x-axis)

```python
# before you start your analysis, you should set your "timestep" column as index-column
df = df.set_index('timestep')
```

```python
# plot the random walks
ax = df.plot()
ax.set(ylabel='x-position', xlabel='time')
```

```python
# plot the positions via histograms
axs = df.hist(bins=50, sharex=True, sharey=True)
axs[1, 0].set(ylabel='frequency')
axs[-1, 0].set(xlabel='x-position')
```

```python
df.describe().T
```

```python
# run some statistics and compare the data
df_describe = df.describe().T
f, ax = plt.subplots(1)
mean = df_describe['mean']
std = df_describe['std']
ax.plot(mean, 'k-o', label='mean')
ax.fill_between(df_describe.index, mean+std, mean-std, label='mean +/- std')
ax.legend()
```

## The central limit walk

* the most important part here is how to look at the data only every nth step
* this is done via `df.iloc[::n]` which uses the general indexing logic of `df.iloc[start:end:step]`
    * so `df.iloc[5:20:3]` would give you the data starting from the 5th entry up to (but not including) the 20th with step length of 3 
        * this corresponds to the timesteps `[5, 8, 11, 14, 17]`

```python
df.iloc[5:20:3]
```

```python
_ = df.iloc[::2].diff().hist(bins=50, sharex=True)
```

```python
_ = df.iloc[::4].diff().hist(bins=50, sharex=True)
```

```python

_ = df.iloc[::16].diff().hist(bins=50, sharex=True)
```

* you observe that the larger the steps get, the closer the step distribution gets to the normal distribution

Explanation:
* taking larger step sizes means for a random walk: add up more random numbers from the same distribution
* so a larger step is a larger sum, and the larger the sum gets, the closer gets the distribution of the "new" random number to the normal distribution
* you observe the central limit theorem and the reason why so many observed distributions are close to a normal distribution
